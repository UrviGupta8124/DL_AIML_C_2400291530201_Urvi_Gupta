```python
#step1
#Input data

# arr1 = list(map(int, input().split()))
# arr2 = list(map(int, input().split()))
arr1=[75, 65, 80]
arr2=[4, 3, 5]
y=1 #Target output: pass=1
b=-1000 #bias
n=0.01 #learning rate
z1=0
z=0 #compute net input
```

```python
#step2: compute net input(z)

for i in range(len(arr1)):
        z1+=arr1[i]*arr2[i]
z=z1+b
print(z)
```

```python
#step3
#Apply Activation Function

if (z<0):
  y_pred=0
elif (z>=0):
  y_pred=1
```

```python
#step4
#Calculate error
error=y-y_pred
print(error)
#error=1 (prediction is wrong->update is required)
```

```python
#step5
#weight update decision/perceptron update rule
import numpy as np

w_new = np.zeros(len(arr1))


for i in range(len(arr1)):
  w_new[i]=arr2[i]+(n*error*arr1[i])
  print("Weight", i, "=", w_new[i])
```

```python
#step6
#bias update
b_new=b+n*error
print(b_new)
```

```python
#repeat step2
z_new=0
z_n=0
for i in range(len(arr1)):
  z_n+=arr1[i]*w_new[i]
z_new=z_n+b_new
print(z_new)
```
